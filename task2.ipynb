```python
import tkinter as tk
from tkinter import messagebox
import random

# Initialize random number and attempts
secret_number = random.randint(1, 100)
attempts = 0

# Function to check guess
def check_guess():
    global attempts
    attempts += 1
    try:
        guess = int(entry.get())
        if guess < 1 or guess > 100:
            message_label.config(text="‚ö†Ô∏è Enter a number between 1 and 100!", fg="orange")
        elif guess < secret_number:
            message_label.config(text=f"‚¨ÜÔ∏è Too low! Attempts: {attempts}", fg="blue")
        elif guess > secret_number:
            message_label.config(text=f"‚¨áÔ∏è Too high! Attempts: {attempts}", fg="red")
        else:
            message_label.config(text=f"üéâ Correct! You guessed in {attempts} attempts!", fg="green")
            messagebox.showinfo("Congratulations!", f"You guessed the number {secret_number} in {attempts} attempts!")
            reset_game()
    except ValueError:
        message_label.config(text="‚ö†Ô∏è Please enter a valid number!", fg="orange")

# Function to reset the game
def reset_game():
    global secret_number, attempts
    secret_number = random.randint(1, 100)
    attempts = 0
    message_label.config(text="Guess a number between 1 and 100", fg="black")
    entry.delete(0, tk.END)

# Function to change background color
def change_bg():
    colors = ["#ffcccc", "#ccffcc", "#ccccff", "#ffffcc", "#ccffff", "#ffccff"]
    root.config(bg=random.choice(colors))
    root.after(1000, change_bg)  # change every 1 second

# GUI setup
root = tk.Tk()
root.title("üéØ Fun Number Guessing Game")
root.geometry("450x300")
root.config(bg="#f0f0f0")

# Start background color animation
change_bg()

# Heading
heading = tk.Label(root, text="üéØ Number Guessing Game", font=("Comic Sans MS", 20, "bold"), bg="#f0f0f0", fg="#4b0082")
heading.pack(pady=10)

# Instruction
instruction = tk.Label(root, text="Guess a number between 1 and 100", font=("Times New Roman", 14), bg="#f0f0f0")
instruction.pack(pady=5)

# Entry
entry = tk.Entry(root, font=("Times New Roman", 16), justify="center")
entry.pack(pady=10)

# Guess button
guess_button = tk.Button(root, text="Guess!", font=("Times New Roman", 14, "bold"), bg="#4b0082", fg="white", command=check_guess)
guess_button.pack(pady=5)

# Reset button
reset_button = tk.Button(root, text="Restart Game", font=("Times New Roman", 14, "bold"), bg="#008080", fg="white", command=reset_game)
reset_button.pack(pady=5)

# Message label
message_label = tk.Label(root, text="Guess a number between 1 and 100", font=("Times New Roman", 12, "bold"), bg="#f0f0f0")
message_label.pack(pady=15)

root.mainloop()
```
